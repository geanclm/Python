```python
# MATRIZ DE EISENHOWER
# by geanclm in 02/05/2022 at 8h
import matplotlib.pyplot as plt
from pylab import plot, show

plt.figure(figsize=(15,5))
plt.xlim(-10,10)
plt.ylim(-10,10)

plt.title(u'MATRIZ DE EISENHOWER', color='blue')
plt.ylabel(u'IMPORTÂNCIA --->', fontsize='large', color='r')
plt.xlabel('URGÊNCIA --->', fontsize='large', color='r')

plt.axhline(y=0, xmin=0.1, xmax=0.95)
plt.axvline(x=0, ymin=0.1, ymax=0.95)

# PRIMEIRO QUADRANTE
plt.annotate('+ IMPORTANTE', xy=(1, 1), xytext=(7.5,8), color='r')
plt.annotate('+ URGENTE', xy=(1, 1), xytext=(7.5,6), color='r')
plt.annotate('EXECUTAR', xy=(1, 1), xytext=(7.5, 4))

eixoXQ1 = [2]
eixoYQ1 = [2]
plot(eixoXQ1, eixoYQ1, marker='o')


# SEGUNDO QUADRANTE
plt.annotate('+ IMPORTANTE', xy=(1, 1), xytext=(-9.5, 8))
plt.annotate('- URGENTE', xy=(1, 1), xytext=(-9.5, 6))
plt.annotate('AGENDAR', xy=(1, 1), xytext=(-9.5, 4))

eixoXQ2 = [-2]
eixoYQ2 = [2]
plot(eixoXQ2, eixoYQ2, marker='o')

# TERCEIRO QUADRANTE
plt.annotate('- IMPORTANTE', xy=(1, 1), xytext=(-9.5, -4), color='orange')
plt.annotate('- URGENTE', xy=(1, 1), xytext=(-9.5, -6), color='orange')
plt.annotate('ELIMINAR', xy=(1, 1), xytext=(-9.5, -8))

eixoXQ3 = [-2]
eixoYQ3 = [-2]
plot(eixoXQ3, eixoYQ3, marker='o')

# QUARTO QUADRANTE
plt.annotate('- IMPORTANTE', xy=(1, 1), xytext=(7.5, -4))
plt.annotate('+ URGENTE', xy=(1, 1), xytext=(7.5, -6))
plt.annotate('DELEGAR', xy=(1, 1), xytext=(7.5, -8))

eixoXQ4 = [2]
eixoYQ4 = [-2]
plot(eixoXQ4, eixoYQ4, marker='o', )

plt.grid(True)
plt.show();
```
