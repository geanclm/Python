```python
# MATRIZ DE EISENHOWER
# by geanclm in 02/05/2022 at 8h
import matplotlib.pyplot as plt

plt.xlim(-9,9)
plt.ylim(-9,9)

plt.title(u'MATRIZ DE EISENHOWER', color='blue')
plt.ylabel(u'IMPORTÂNCIA --->', fontsize='large', color='r')
plt.xlabel('URGÊNCIA --->', fontsize='large', color='r')

plt.axhline(y=0, xmin=0.1, xmax=0.9)
plt.axvline(x=0, ymin=0.1, ymax=0.9)

# PRIMEIRO QUADRANTE
plt.annotate('+ IMPORTANTE', xy=(1, 1), xytext=(4, 7), color='r')
plt.annotate('+ URGENTE', xy=(1, 1), xytext=(4, 5), color='r')
plt.annotate('EXECUTAR', xy=(1, 1), xytext=(4, 3))

# SEGUNDO QUADRANTE
plt.annotate('+ IMPORTANTE', xy=(1, 1), xytext=(-7, 7))
plt.annotate('- URGENTE', xy=(1, 1), xytext=(-7, 5))
plt.annotate('AGENDAR', xy=(1, 1), xytext=(-7, 3))

# TERCEIRO QUADRANTE
plt.annotate('- IMPORTANTE', xy=(1, 1), xytext=(-7, -3), color='orange')
plt.annotate('- URGENTE', xy=(1, 1), xytext=(-7, -5), color='orange')
plt.annotate('ELIMINAR', xy=(1, 1), xytext=(-7, -7))

# QUARTO QUADRANTE
plt.annotate('- IMPORTANTE', xy=(1, 1), xytext=(4, -3))
plt.annotate('+ URGENTE', xy=(1, 1), xytext=(4, -5))
plt.annotate('DELEGAR', xy=(1, 1), xytext=(4, -7))

# plt.grid(True)
plt.show();
```
